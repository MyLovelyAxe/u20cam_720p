```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from constants import CALIB_PARAM_JSON
from camera.camera import Camera, Extrinsics
from camera.projection import Projector, Rotation
```

#### Load original image

```python
ori_img = cv2.imread("undistorted.png")
ori_img = cv2.cvtColor(ori_img, cv2.COLOR_BGR2RGB)
print(ori_img.shape)
plt.imshow(ori_img)
```

#### Create source camera

```python
src_cam = Camera.create_from_json(CALIB_PARAM_JSON)
src_cam.extrinsics = Extrinsics(
    rotation = Rotation(alpha=0, beta=0, gamma=0).R,
    translation = np.array([0,0,0]),
)
```

#### Rotate the source camera for target camera

```python
rotation = Rotation(alpha=0, beta=-15, gamma=0) # rotate 15° to the right
tgt_cam = rotation.rotate_camera(src_camera=src_cam)
```

#### Create projector

```python
projector = Projector(src_cam=src_cam, tgt_cam=tgt_cam)
```

#### Project original image

```python
projected_img = projector(img=ori_img)
print(projected_img.shape)
plt.imshow(projected_img)
```
